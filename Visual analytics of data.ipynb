```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('kc-house-data.csv')
df.sample(n=21)
```

```python
plt.figure(figsize=(8, 6))
plt.plot([0, 1, 2, 3], [10, 9, 6, 1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example')
plt.grid();
```

**Мы наблюдаем в этой гистрограмме то что большой всплеск построек был после 2000 года и после 2015 года наблюдается просадок. Меньше всего домов строилось в период 1930-1940 годы возможно из-за войны**

```python
plt.figure(figsize=(6, 4))

sns.histplot(df['yr_built'], bins=20)

plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Кол-во паостроенных домов')
```

**По этой гистограмме мы видим что больше всего домов строилось с квадратурой в разрзе от 1000 до чуть больше 2000 кв.м, также наблюбдаем что есть незначительное кол-во построек до 8000 кв.м и чуть больше**

```python
plt.figure(figsize=(6, 4))

sns.histplot(df['sqft_living'], bins=20)

plt.title('Распрделение жилой квадратуры')
plt.xlabel('Жилая квадатура')
plt.ylabel('Кол-во паостроенных домов')
```

```python
data = df.groupby('yr_built')['price'].sum().reset_index()
data.head()
```

**По нижеуказанномму графику разброса мы наблюдаем что цена весьма сильна повысилась после 2015 года постройки и есть также падения цены после этого периода, возможно был экономический кризис или сопутствующие факторы**

```python
plt.figure(figsize=(6, 4))

plt.scatter(data['yr_built'], data['price'])

plt.title('Динамика изменения цены в зависимости от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Цена');
```

**Этот ящик с усами подтверждат нам что домов с видом на набержную не наблюдалось**

```python
plt.figure(figsize=(6, 3))

sns.boxplot(x=df['waterfront'], whis=1.5)

plt.xlabel('waterfront')
plt.show()
```

**По этой диаграмме распределения мы видим что больше всего домов было с меньшими этажами, а минимум построек мы наблюдаем домов где была больше 2 и меньше 3 этажей**

```python
plt.figure(figsize=(8, 6))
plt.hist(df['floors'])
plt.title('Распределение этажей домов')
plt.xlabel('Этажи')
plt.ylabel('Количество');
```

**Диаграмма которая указан наже свидетельствует нам о том что дома строились весьма хорошие, потому что наибольшее кол-во домов строились с состоянием больше среднего бала**

```python
plt.figure(figsize=(8, 6))

plt.hist(df['condition'])

plt.title('Диаграмма распределения состояния домов')
plt.xlabel('Состояние')
plt.ylabel('Количество');
```

**Ну и на последок мы перечислим особенности которые влияют на цену домов, это квадратура чем она выще тем цена выше, также это этажность дома, вид на набережную, состояние дома и год его постройки**
